```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Social_Network_Ads.csv')
```

```python
df.head()
```

```python
purchased = df[df['Purchased'] == 1]
not_purchased = df[df['Purchased'] == 0]
```

```python
plt.scatter(not_purchased['EstimatedSalary'], not_purchased['Age'], label='no comprado', color='blue')
plt.scatter(purchased['EstimatedSalary'], purchased['Age'], label='comprado', color='red')

# Adicionar rtulos aos eixos e ttulo
plt.xlabel('Salrio')
plt.ylabel('Idade')
plt.title('Grfico de Disperso - Idade/Salrio x Quantidade de Compra')

# Adicionar legenda
plt.legend()

# Exibir o grfico
plt.show()
```
