```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Social_Network_Ads.csv')
```

```python
df.head()
```

```python
purchased = df[df['Purchased'] == 1]
not_purchased = df[df['Purchased'] == 0]
```

```python
plt.scatter(not_purchased['EstimatedSalary'], not_purchased['Age'], label='não comprado', color='blue')
plt.scatter(purchased['EstimatedSalary'], purchased['Age'], label='comprado', color='red')

# Adicionar rótulos aos eixos e título
plt.xlabel('Salário')
plt.ylabel('Idade')
plt.title('Gráfico de Dispersão - Idade/Salário x Quantidade de Compra')

# Adicionar legenda
plt.legend()

# Exibir o gráfico
plt.show()
```
