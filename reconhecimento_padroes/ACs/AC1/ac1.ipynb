```python
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

> Função para exibir imagens

```python
def display_image(image, text='Imagem'):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(text)
    plt.axis('off')
    plt.show()
```

```python
img_original = cv2.imread('1-Avioes.webp')
img = cv2.imread('1-Avioes.webp')
display_image(img)
```

> Transformar em cinza

```python
img_cinza = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display_image(img_cinza)
```

> Invertendo as escolas de cinza

```python
imagem_invertida = cv2.bitwise_not(img_cinza)
display_image(imagem_invertida)
```

> aplicação de blur para remover ruidos e auxiliar a identificar bordas

```python
img_blur = cv2.GaussianBlur(imagem_invertida, (11, 11), 0)
display_image(img_blur)
```

> Binarização

```python
_, img_binary = cv2.threshold(img_blur, 240, 255, cv2.THRESH_BINARY)
display_image(img_binary, 'Imagem binarizada')
```

> Dectação de bordas

```python
bordas = cv2.Canny(img_binary, 70, 120)
display_image(bordas)
```

```python
contornos, hierarquia = cv2.findContours(bordas, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(img, contornos, -1, (0, 255, 0), 2)

print('Contornos encontrados: ' + str(len(contornos)))
display_image(img)
```

```python
display_image(img_original, "Imagem Original")
display_image(img_cinza, "Imagem com escalas de cinza")
display_image(imagem_invertida, "Imagem com escala de cinza invertida")
display_image(img_blur, "Imagem com blur")
display_image(img_binary, "Imagem binarizada")
display_image(bordas, "Aplicando as bordas")
print('Contornos encontrados: ' + str(len(contornos)))
display_image(img, "Imagem original contornada")

```
