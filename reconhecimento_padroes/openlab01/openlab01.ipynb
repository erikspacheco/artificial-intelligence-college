```python
from matplotlib import pyplot as plt
import cv2 

img = cv2.imread('gato.jpg')
# cv2.imshow('Imagem RGB', img)

canais = cv2.split(img)
color = ("b", "g", "r")

plt.figure()
plt.title('Histograma RGB')
plt.xlabel('Intensidade')
plt.ylabel('Qtde de Pixeis')

for (canal, cor) in zip(canais, color):
    hist = cv2.calcHist([canal], [0], None, [256], [0, 256])
    plt.plot(hist, color = cor)
    plt.xlim([0, 256])

plt.show()

cv2.waitKey(27)
```

```python
from matplotlib import pyplot as plt
import cv2 

img = cv2.imread('gato.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# cv2.imshow('Imagem Tons de Cinza', img)

hist = cv2.calcHist([img], [0], None, [256], [0, 256])

plt.figure()
plt.title('Histograma Cinza')
plt.xlabel('Intensidade')
plt.ylabel('Qtde de Pixeis')
plt.plot(hist, color = 'gray')
plt.xlim([0, 256])

plt.show()

cv2.waitKey(27)
```

```python
# Constraste em imagem cinza

from matplotlib import pyplot as plt
import numpy as np
import cv2

imagem = cv2.imread('gato.jpg')
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
constraste = cv2.equalizeHist(imagem)

# cv2.imshow("Imagem Equalizada", constraste)
plt.figure()
plt.title("Histograma Equalizado")
plt.xlabel("Intensidade")
plt.ylabel("Qtde. de Pixeis")
plt.hist(constraste.ravel(), 256, [0, 256])
plt.xlim([0, 256])
plt.show()
           

# cv2.imshow("Image Original", imagem)
plt.figure()
plt.title("Histograma Original")
plt.xlabel("Intensidade")
plt.ylabel("Qtde. de Pixeis")
plt.hist(imagem.ravel(), 256, [0, 256])
plt.xlim([0, 256])
plt.show()

cv2.waitKey(27)
```
