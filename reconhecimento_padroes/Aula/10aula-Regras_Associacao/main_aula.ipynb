# Regras de Associação

```python
import numpy as np
import matplotlib as plt
import pandas as pd
from apyori import apriori
```

```python
df = pd.read_csv('store_data.csv', delimiter= ',', header= None)
df.head(10)
```

```python
# Tratar os dados tirando NAN e incluindo 0 no lugar
df.replace(np.nan, 0, inplace=True)
df.head(10)
```

```python
def removeZeros(lista):
    return list(filter(lambda x: x != 0, lista))
```

```python
listao = []
for index, product in df.iterrows():
    lista = product.values.tolist()
    lista = removeZeros(lista)
    listao.append(lista)
    # print(index, product)
    print(lista)
```

```python
regras = apriori(transactions= listao, min_support = 0.002, min_confidence = 0.02, min_lift = 2)
```

```python
final = list(regras)
final
```

```python
# Tratar usando tuplas (tuple)

def fim (final):
    X = [tuple(result[2][0][0])[0] for result in final]
    Y = [tuple(result[2][0][1])[0] for result in final]
    sup = [result[1] for result in final]
    conf = [result[2][0][2] for result in final]
    lif = [result[2][0][3] for result in final]
    return list (zip(X, Y, sup, conf, lif))

```

```python
resultDF = pd.DataFrame(fim(final), columns = ['X', 'Y', 'Suporte', 'Confianca', 'Lift'])
```

```python
resultDF.sort_values(by='Confianca', ascending=False).head(50)
```

# ML EXTEND

mesma ideia, porém usa um modelo já treinado, no caso uma ML

```python
import numpy as np
import matplotlib as plt
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('store_data.csv', delimiter= ',', header= None)
df.replace(np.nan, 0, inplace=True)
df.head(10)
```

```python
def removeZeros(lista):
    return list(filter(lambda x: x != 0, lista))
```

```python
listao = []
for index, product in df.iterrows():
    lista = product.values.tolist()
    lista = removeZeros(lista)
    listao.append(lista)
```

```python
TE = TransactionEncoder()
TE_AR = TE.fit(listao).transform(listao)
TE_AR
```

```python
df = pd.DataFrame(TE_AR, columns=TE.columns_)
df.head(25)
```

```python
#  Teste sem a coluna água
# df.drop('mineral water', axis=1, inplace=True) 
# df.head(25)
```

```python
freq_itemset = apriori(df, min_support= 0.01, use_colnames=True)
freq_itemset.sort_values(by=['support'], ascending= False).head(25)
```

```python
regras = association_rules(freq_itemset, metric = 'confidence', min_threshold = 0.4)
# regras.sort_values(by= 'confidence', ascending = False)
regras.sort_values(by= 'confidence', ascending = False).drop(columns=['antecedent support', 'consequent support', 'leverage', 'zhangs_metric'], axis = 1) 
```
